```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.output_parsers import (
    JsonOutputParser,
    CommaSeparatedListOutputParser,
)
from google.colab import userdata
from rich.pretty import pprint
import os

# è¨­å®šAPI Key
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
# å»ºç«‹æ¨¡å‹
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

### CommaSeperatedListOutputParser

```python
# æ¨¡å‹ç”Ÿæˆçš„æ–‡æœ¬ï¼šä¸€å€‹ä»¥é€—è™Ÿåˆ†éš”çš„åˆ—è¡¨
model_output = "apple, orange, banana, grape"

# å‰µå»º CommaSeparatedListOutputParser è§£æå™¨
parser = CommaSeparatedListOutputParser()

# è§£ææ¨¡å‹çš„è¼¸å‡º
parsed_output = parser.parse(model_output)

# é¡¯ç¤ºè§£æå¾Œçš„çµæœ
print(parsed_output, type(parsed_output), len(parsed_output))  # è¼¸å‡ºï¼š['apple', 'orange', 'banana', 'grape']
```
